```python
from pandas import DataFrame
Data = {'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
	      'y': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]}

df = DataFrame(Data,columns=['x','y'])   #asigno a dataframe   
print(df)
```

```python
import matplotlib.pyplot as plt  #libreria de grafico
plt.scatter(df['x'],df['y'])      #con la grafica de dispercion asigno valores a mostrar
plt.show()                        #imprimo grafica
```

```python
from sklearn.cluster import KMeans    # ubico los centroides existentes dentro de la grafica
kmeans = KMeans(n_clusters=5).fit(df)  #me muestra 3 centroides que es lo que parametrizo, se puede verificar con mas centroides
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5) #busca centroides y los configuro en amarillo
plt.scatter(centroids[:,0], centroids[:,1], c='orange', s=50)
plt.show
```
